```python
#Importing the Essential libraries and their Functions

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import nltk
import re
```

```python
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("/content/potential-talents.csv")

data
```

```python
data.drop(['id', 'location', 'fit'], axis=1, inplace=True)

data
```

```python
data_2 = data.copy()

data_2
```

```python
data_2.dropna(subset=['job_title'], inplace=True)

data_2
```

```python
pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer

data_2['job_title'] = data_2['job_title'].astype(str)

# Load Pretrained MiniLM Model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')  # 384D embeddings

# Generate Sentence Embeddings
data_2['job_title_embedding'] = data_2['job_title'].apply(lambda x: model.encode(x))

# Display Results
print(data_2[['job_title', 'job_title_embedding']])
```

```python
w1 = model.encode('Aspiring human resources')
```

```python
data_2['cosine_similarity'] = data_2['job_title_embedding'].apply(lambda x: cosine_similarity([w1], [x])[0][0])

data_2['cosine_similarity']
```

```python
data_2
```

```python
data_2 = data_2[['job_title', 'job_title_embedding', 'cosine_similarity', 'connection']]

data_2
```

```python
data_3 = data_2.copy()
```

```python
data_3['connection'] = data_2['connection'].replace(['500+ '], [500], inplace=True)
```

```python
data_2['connection'] = data_2['connection'].astype(int)

data_2['connection']
```

```python
ranked_data = data_2.copy()

ranked_data
```

```python
ranked_data.drop(['job_title', 'job_title_embedding', 'cosine_similarity'], axis=1, inplace=True)

ranked_data
```

#Feature Scaling

```python
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(ranked_data)
ranked_data['scaled_connection'] = pd.DataFrame(scaled_data, columns=ranked_data.columns)

ranked_data['scaled_connection']
```

```python
ranked_data
```

```python
ranked_data[['job_title', 'job_title_embedding', 'cosine_similarity']] = data_2[['job_title', 'job_title_embedding', 'cosine_similarity']]

ranked_data
```

```python
ranked_data.drop(['connection'], axis=1, inplace=True)

ranked_data
```

```python
ranked_data = ranked_data[['job_title', 'job_title_embedding', 'cosine_similarity', 'scaled_connection']]

ranked_data
```

#Ranking

```python
ranked_data['ranking'] = 0.75 * ranked_data['cosine_similarity'] + 0.25 * ranked_data['scaled_connection']

ranked_data['ranking']
```

```python
ranked_data
```

#Sorting (Descending)

```python
ranked_data = ranked_data.sort_values(by='ranking', ascending=False)

ranked_data['ranking']
```
